```python
from pymysql import Connection
class Record:
    date: str = None
    order_id: str = None
    money: int = None
    province: str = None

    def __init__(self, date: str, order_id: str, money: int, province: str):
        self.date = date
        self.order_id = order_id
        self.money = money
        self.province = province

    def __str__(self):
        return f"[{self.date},{self.order_id},{self.money},{self.province}]"

    __repr__ = __str__
conn=Connection(
    host="localhost",
    port=3306,
    user="root",
    password="123456",
)
conn.select_db("test")
cursor=conn.cursor()
cursor.execute("select * from sale")
result=cursor.fetchall()
f=open("file/sql_test.txt","w",encoding="UTF-8")
for data in result:
    data_dict=dict()
    keys=["data","order_id","money","province"]
    j=0
    for i in keys:
        data_dict[i]=data[j]
        j+=1
    data_json=json.dumps(data_dict,ensure_ascii=False)
    f.write(data_json+'\n')
f.close()
conn.close()
```
