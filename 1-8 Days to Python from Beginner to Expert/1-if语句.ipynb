```python
import random
num=random.randint(1,10)
res=int(input("请猜一个数："))
if res==num:
    print("恭喜，第一次就猜对了！")
else:
    if res<num:
        print("第一次猜小了！")
    else:
        print("第一次猜大了！")
    res=int(input("请猜一个数："))
    if res==num:
        print("恭喜，第二次就猜对了！")
    else:
        if res<num:
            print("第二次猜小了！")
        else:
            print("第二次猜大了！")
        res=int(input("请猜一个数："))
        if res==num:
            print("恭喜，第三次就猜对了！")
        else:
            if res<num:
                print("第三次猜小了！")
            else:
                print("第三次猜大了！")
```
