```python
import json
from pyecharts.charts import Line
from pyecharts.options import LabelOpts,TitleOpts
f_us=open("file/line/美国.txt","r",encoding="UTF-8")
us_data=f_us.read()
f_jp=open("file/line/日本.txt","r",encoding="UTF-8")
jp_data=f_jp.read()
f_in=open("file/line/印度.txt","r",encoding="UTF-8")
in_data=f_in.read()

us_data=us_data.replace("jsonp_1629344292311_69436(","")
us_data=us_data[:-2]
jp_data=jp_data.replace("jsonp_1629350871167_29498(","")
jp_data=jp_data[:-2]
in_data=in_data.replace("jsonp_1629350745930_63180(","")
in_data=in_data[:-2]

us_dict=json.loads(us_data)
us_trend=us_dict["data"][0]["trend"]
us_x=us_trend["updateDate"][:314]
us_y=us_trend["list"][0]["data"][:314]

jp_dict=json.loads(jp_data)
jp_trend=jp_dict["data"][0]["trend"]
jp_x=jp_trend["updateDate"][:314]
jp_y=jp_trend["list"][0]["data"][:314]

in_dict=json.loads(in_data)
in_trend=in_dict["data"][0]["trend"]
in_x=in_trend["updateDate"][:314]
in_y=in_trend["list"][0]["data"][:314]

line=Line()
line.add_xaxis(us_x)
line.add_yaxis("美国确诊人数",us_y,label_opts=LabelOpts(is_show=False))
line.add_yaxis("日本确诊人数",jp_y,label_opts=LabelOpts(is_show=False))
line.add_yaxis("印度确诊人数",in_y,label_opts=LabelOpts(is_show=False))


line.set_global_opts(
    title_opts=TitleOpts(title="2020年疫情确诊人数折线图",pos_left="center",pos_top="5%"),

)
line.render()

f_us.close()
f_jp.close()
f_in.close()
```

```python
from pyecharts.charts import Bar,Timeline
import pyecharts.options as opts
from pyecharts.globals import ThemeType
bar1 = Bar(
        # ✅ 关键：子图也要设置相同主题
        init_opts=opts.InitOpts(theme=ThemeType.LIGHT)
    )
bar1.add_xaxis(["中国","美国","日本"])
bar1.add_yaxis("GDP",[1000,800,700],label_opts=opts.LabelOpts(position="right"))
bar1.reversal_axis()
bar1.set_global_opts(
    legend_opts=opts.LegendOpts(
        pos_top="5%",     # 距离顶部 5%
        pos_left="center" # 水平居中
    ),
    xaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False)),
    yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False))
)

bar2 = Bar(
        # ✅ 关键：子图也要设置相同主题
        init_opts=opts.InitOpts(theme=ThemeType.LIGHT)
    )
bar2.add_xaxis(["中国","美国","日本"])
bar2.add_yaxis("GDP",[700,600,500],label_opts=opts.LabelOpts(position="right"))
bar2.reversal_axis()
bar2.set_global_opts(
    legend_opts=opts.LegendOpts(
        pos_top="5%",     # 距离顶部 5%
        pos_left="center" # 水平居中
    ),
    xaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False)),
    yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False))
)

bar3 = Bar(
        # ✅ 关键：子图也要设置相同主题
        init_opts=opts.InitOpts(theme=ThemeType.LIGHT)
    )
bar3.add_xaxis(["中国","美国","日本"])
bar3.add_yaxis("GDP",[500,400,300],label_opts=opts.LabelOpts(position="right"))
bar3.reversal_axis()
bar3.set_global_opts(
    legend_opts=opts.LegendOpts(
        pos_top="5%",     # 距离顶部 5%
        pos_left="center" # 水平居中
    ),
    xaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False)),
    yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False))
)

# ✅ 修复：主题必须用 init_opts=opts.InitOpts() 包裹
timeline = Timeline(
    init_opts=opts.InitOpts(theme=ThemeType.LIGHT)
)
timeline.add(bar3,"2020")
timeline.add(bar2,"2023")
timeline.add(bar1,"2026")
timeline.add_schema(
    is_auto_play=True,
    is_loop_play=True,
    play_interval=1000,
    is_timeline_show=True,

)
timeline.render()
```

```python
from pyecharts.charts import Bar,Timeline
from pyecharts.options import *
from pyecharts.globals import ThemeType
f=open("F:/install/tool/python/heima/资料/动态柱状图数据/1960-2019全球GDP数据.csv","r",encoding="GB2312")
data_lines=f.readlines()
data_lines.pop(0)
data_dict=dict()
for data_line in data_lines:
    year=int(data_line.split(",")[0])
    country=data_line.split(",")[1]
    GDP=float(data_line.split(",")[2])/100000000
    try:
        data_dict[year].append([country,GDP])
    except KeyError:
        data_dict[year]=[]
        data_dict[year].append([country,GDP])
years_key=data_dict.keys()
sorted(years_key)

timeline=Timeline({"theme":ThemeType.LIGHT})
for year in years_key:
    data_dict[year].sort(key=lambda element:element[1],reverse=True)
    year_data=data_dict[year][:8]
    year_data=year_data[::-1]
    data_x=[]
    data_y=[]
    for c_G in year_data:
        data_x.append(c_G[0])
        data_y.append(c_G[1])
    bar=Bar()
    bar.add_xaxis(data_x)
    bar.add_yaxis("GDP(亿)",data_y,label_opts=LabelOpts(position="right"))
    bar.reversal_axis()
    bar.set_global_opts(
    title_opts=TitleOpts(title=f"{year}年全球GDP前八",pos_left="left"),
    legend_opts=opts.LegendOpts(
        pos_top="1%",     # 距离顶部 5%
        pos_left="center" # 水平居中
    ),
    xaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False)),
    yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=False))
)
    timeline.add(bar,str(year))

timeline.add_schema(
    is_auto_play=True,
    is_loop_play=True,
    play_interval=500,
    is_timeline_show=True,

)
timeline.render()
```
