```python
class Phone:
    __is_5g_enable=False
    def __check_5g(self):
        if self.__is_5g_enable:
            print("5g启动")
        else:
            print("5g关闭，使用4g网络")
    def call_by_5g(self):
        self.__check_5g()
        print("正在通话")

class Phone2022(Phone):
    type="2022"
class NFC:
    type="NFC"
    def read(self):
        print(f"{self.type}读卡中")
    def write(self):
        print("写卡中")
class RemoteCon:
    type="红外线"
    def con(self):
        print("红外线启动")
class MyPhone(Phone,NFC,RemoteCon):
    pass
phone=MyPhone()
phone.call_by_5g()
phone.read()
phone.con()
```
