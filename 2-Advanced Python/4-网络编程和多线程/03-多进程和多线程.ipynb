```python
import multiprocessing
import time
def coding():
    for i in range(1,11):
        print(f"正在打印第{i}行代码！")
        time.sleep(0.1)
def music():
    for i in range(1,11):
        print(f"正在听第{i}遍音乐。。。")
        time.sleep(0.1)
p1=multiprocessing.Process(target=coding)
p2=multiprocessing.Process(target=music)
p1.start()
p2.start()
```
