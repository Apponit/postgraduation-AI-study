```python
# 1.定义装饰器
def hint(func):
    # 1.1定义闭包函数
    def inner():
        print("正在努力计算中....")
        result = func()
        print(f"计算结果为: {result}")
        return result
    # 1.2返回闭包函数
    return inner

# 装饰器的语法糖写法
@hint
# 2.定义被装饰的函数（无参数，有返回值）
def sum_num():
    a=10
    b=20
    return a+b

# 调用函数并接收返回值
result = sum_num()
print(f"最终结果: {result}")
```

```python
# 1.定义装饰器
def hint(func):
    # 1.1定义闭包函数
    def inner():
        print("正在努力计算中....")
        func()
    # 1.2返回闭包函数
    return inner
# 装饰器的语法糖写法
@hint
# 2.定义被装饰的函数
def sum_num():
    a=10
    b=20
    print(a+b)
sum_num()
```

```python
# 1.定义装饰器
def my_decorator(func):
    # 1.1定义闭包函数
    def fn_inner(a:int,b:int):
        print("正在努力计算中....")
        func(a, b)
    # 1.2返回闭包函数
    return fn_inner
# 装饰器的语法糖写法
@my_decorator
# 2.定义被装饰的函数
def sum_num(a:int,b:int):
    print(a+b)
# 3.调用函数
sum_num(10,20)
```
