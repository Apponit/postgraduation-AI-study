```python
class Student:
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight

    def __str__(self):
        return f"{self.name}的体重是:{self.weight}kg"

    def workout(self):
        self.weight -= 0.5

    def eat(self):
        self.weight += 2


xiaoming = Student("小明", 100)
print(xiaoming)
xiaoming.eat()
print(xiaoming)
xiaoming.workout()
print(xiaoming)
```

```python
class SweetPotato:
    def __init__(self):
        self.cook_time:int = 0
        self.cook_state:str = ""
        self.tl:list[str] = []

    def __str__(self):
        return f"{self.cook_time}+{self.cook_state}+{self.tl}"

    def cook(self, time):
        if time<=0:
            print("无效值")
            return
        self.cook_time = time
        if 0 < time < 3:
            self.cook_state="生的"
        elif 3<time<7:
            self.cook_state="半生不熟"
        elif 7<time<12:
            self.cook_state="熟了"
        else:
            self.cook_state="已烤焦,糊了"
    def add_tl(self,tl:str):
        self.tl.append(tl)
m1=SweetPotato()
m1.cook(8)
m1.add_tl("surgy")
m1.add_tl("paper")
print(m1)

```
