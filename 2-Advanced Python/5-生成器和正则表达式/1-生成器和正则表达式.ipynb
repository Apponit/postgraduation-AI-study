```python
from fontTools.afmLib import readlines


# 1.自定义迭代器类
class MyIterator(object):
    # 1.1通过init魔法方法，初始化属性，指定范围
    def __init__(self,start,end):
        self.current_value=start
        self.end=end
    # 1.2 重写interator方法，返回当前对象（迭代器对象）
    def __iter__(self):
        return self
    # 1.3 重写next方法，返回当前值，并更新当前值
    def __next__(self):
        # 1.3.1 判断越界异常，并抛出异常
        if self.current_value>=self.end:
            raise StopIteration
        self.current_value+=1
        return self.current_value-1
if __name__ == '__main__':
    # 2.测试
    # 2.1 for循环测试
    for i in MyIterator(1,8):
        print(i)
    print("*"*23)

    # 2.2 next函数测试
    iterator=MyIterator(10,18)
    print(next(iterator))
```

```python
import math
# 案例从歌词文件中获得指定个数的歌词数
def my_gt1(batch_size):
    """
    读取文件，根据指定的batch_size返回生成器
    :param batch_size: 指定的歌词数
    :return: 生成器
    """
    with open("./data/geci.txt","r",encoding="utf-8") as f:
        data=[line.strip() for line in f.readlines()]
        total_batch=math.ceil(len(data)/batch_size)
        for i in range(total_batch):
            yield data[i*batch_size:(i+1)*batch_size]


```

```python
class Student(object):
    def __init__(self,age):
        self.__age=age
    # 对于私有属性，设置get 和set方法，并设置property属性，一般get和set方法和私有属性的名字相同
    @property
    def age(self):
        """
        get_age
        :return:
        """
        return self.__age
    @age.setter
    def age(self,age):
        """
        set_age
        :param age:
        :return:
        """
        self.__age=age
if __name__ == '__main__':
    s=Student(10)
    s.age=20
    print(s.age)
```
