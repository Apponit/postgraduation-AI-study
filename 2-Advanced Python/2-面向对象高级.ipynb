```python
class Master(object):
    def __init__(self):
        self.kong_fu="古法制造煎饼果子"
        print(id(self))
    def make_cake(self):
        print(f"{self.kong_fu}----")
class School(Master):
    def __init__(self):
        self.kong_fu="大众方法制造煎饼果子"
    def make_cake(self):
        print(f"{self.kong_fu}----")
class Prentice(School,Master):
    def __init__(self):
        self.kong_fu="独创方法制造煎饼果子"
    def make_cake(self):
        print(f"{self.kong_fu}---")
    def make_master_cake(self):
        print(id(self))
        Master.make_cake(self)
        Master.__init__(self)
        Master.make_cake(self)
        print("-"*10)
        print(self.kong_fu)


if __name__ == '__main__':
    p=Prentice()
    p.make_cake()
    p.make_master_cake()
```

```python
class Animal(object):
    def speak(self):
        pass
class Dog(Animal):
    def speak(self):
        print("汪汪汪")

class Cat(Animal):
    def speak(self):
        print("喵喵喵")
def an_speak(an:Animal):
    an.speak()
dog=Dog()
an_speak(dog)
cat=Cat()
an_speak(cat)
```

```python
class HeroFighter:
    def get_value(self):
        return 60
class AdvHeroFight:
    def get_value(self):
        return 100
class DiFighter:
    def get_value(self):
        return 70
def fight_state(hero:HeroFighter,di:DiFighter):
    if hero.get_value()>=di.get_value():
        print("hero win!")
    else:
        print("di win!")
if __name__ == '__main__':
    h1=HeroFighter()
    h2=AdvHeroFight()
    d1=DiFighter()
    fight_state(h1,d1)
    fight_state(h2,d1)
```
