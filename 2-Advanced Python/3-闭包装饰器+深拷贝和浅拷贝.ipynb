```python
from ipaddress import summarize_address_range


# 1.定义装饰器
def hint(func):
    # 1.1定义闭包函数
    def inner():
        print("正在努力计算中....")
        result = func()
        print(f"计算结果为: {result}")
        return result

    # 1.2返回闭包函数
    return inner


# 装饰器的语法糖写法
@hint
# 2.定义被装饰的函数（无参数，有返回值）
def sum_num():
    a = 10
    b = 20
    return a + b


# 调用函数并接收返回值
result = sum_num()
print(f"最终结果: {result}")
```

```python
# 1.定义装饰器
def my_decorator(func):
    # 1.1定义闭包函数
    def fn_inner(a: int, b: int):
        print("正在努力计算中....")
        func(a, b)

    # 1.2返回闭包函数
    return fn_inner


# 装饰器的语法糖写法
@my_decorator
# 2.定义被装饰的函数
def sum_num(a: int, b: int):
    print(a + b)


# 3.调用函数
sum_num(10, 20)
```

```python
def fn_decorator(func):
    def fn_inner(a, b):
        print("正在努力计算中....")
        res = func(a, b)
        print("计算结束")
        return res

    return fn_inner


@fn_decorator
def sum_num(a, b):
    return a + b


# sum_num=fn_decorator(sum_num)
print(sum_num(10, 20))
```

```python
# 1.定义装饰器
def loging(flag):
    # 2.定义嵌套的装饰器（外面装饰器的内部函数）
    def decorator(func):
        # 3.定义内部函数
        def inner(a,b):
            # 3.1添加额外功能
            if flag=="-":
                print("正在减法运算")
                res=func(a,b)
            else:
                print("正在加法运算")
                res=func(a,b)
            # inner是内部函数，要和被修饰的原函数保持一致，有返回值
            return res
        # 2.1返回内部函数
        return inner
    # 1.1返回嵌套的装饰器，（外部装饰器的内部函数）
    return decorator
@loging("+")
def add(a,b):
    return a+b
@loging("-")
def sub(a,b):
    return a-b
print(add(1,2))
print("-"*32)
print(sub(1,2))
```

```python
def decorator(func):
    def inner(a,b):
        if func.__name__=="add":
            print("正在计算加法")
            return func(a,b)
        else:
            print("正在计算减法")
            return func(a,b)
    return inner


@decorator
def add(a,b):
    return a+b
@decorator
def sub(a,b):
    return a-b
print(add(1,2))
print(sub(1,2))
```
